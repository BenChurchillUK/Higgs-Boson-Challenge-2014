## The ATLAS Higgs Boson Challenge (2014) Data

This notebook serves as an Exploratory Data Analysis, to understand the dataset and what preprocessing is required.

### Import Packages and Modules

```python
import pandas as pd
from pathlib import Path
```

### Pull Data

```python
root = Path().resolve()
while not (root/"src").exists():
    root = root.parent

df = pd.read_csv(root/"data"/"atlas-higgs-challenge-2014-v2.csv")
```

### Dataset Columns

```python
print(df.columns)
```

### Shape of the Dataset

To identify the number of rows (records) and columns (features) in the dataset.

```python
print(df.shape)
```

### Dataset Information

The important features to note are 'Label' (str) and 'PRI_jet_num' (int64). All other relevant features are float64.

```python
print(f"{df.info()}")
```

### Description of the Dataset

Describe identifies key statistics of each feature, such as the mean and standard deviation.

```python
print(df.describe())
```

### Null Values

While info() gives null values, it is worthwhile reconfirming incase any further preprossesing needs to be done. In this case, there were no NaN values.

```python
print(df.isnull().sum())
```

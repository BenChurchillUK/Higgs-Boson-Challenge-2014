### Libraries, Modules and Data

```python
import matplotlib.pyplot as plt
import pandas as pd
from pathlib import Path

root = Path().resolve()
while not (root/"src").exists():
    root = root.parent

df = pd.read_csv(root/"data"/"atlas-higgs-challenge-2014-v2.csv")
signal = df[df["Label"] == "s"]
noise = df[df["Label"] == "b"]
```
